# 1. Définition du webscraping

Le web scraping est une technique utilisée pour extraire des données à partir de sites web. Cela se fait généralement en utilisant des scripts ou des programmes automatisés qui parcourent les pages web et récupèrent les informations souhaitées. Un scraper web extrait le code HTML d'une page et les données stockées dans sa base de données sous-jacente, puis les exporte vers un tiers.

Les **cas d'usage** courants du web scraping peuvent être :

1. Faire des analyses sur le marché en collectant les données sur les prix, les produits et les avis des consommateurs pour surveiller la concurrence et identifier les tendances du marché.
2. Rarassembler des informations provenant de différentes sources pour créer des bases de données ou des portails d'information.
3. Extraire des données pour des études et des analyses dans divers domaines de recherche.
4. Collecter des données d'un réseau social sur les mentions, les hashtags et les tendances pour analyser l'engagement et la perception de la marque.
5. Automatiser la collecte de données répétitives pour gagner du temps.

# 4. Ecrire un script python pour récupérer l’ensemble des données de la page

a. D’abord une page de description détaillée de votre solution

Pour récupérer l'ensemble des données de la page du site Quotes to Scrape, nous allons utiliser la bibliothèque BeautifulSoup en Python. Voici les étapes détaillées de notre solution :

**Envoyer une requête HTTP :** Utiliser la bibliothèque requests pour envoyer une requête HTTP à la page cible et récupérer le contenu HTML.

Parser le contenu HTML **texte en gras**: Utiliser BeautifulSoup pour parser le contenu HTML et naviguer dans la structure du document.

**Extraire les données :** Identifier les balises HTML contenant les données souhaitées (citations, auteurs, tags) et extraire ces informations.

**Stocker les données :** Organiser les données extraites dans un format structuré, comme une liste de dictionnaires, et les afficher ou les enregistrer dans un fichier.

b. Script Python

```python
import requests
from bs4 import BeautifulSoup

def recuperer_donnees(url):
    # Envoyer une requête HTTP à la page cible
    response = requests.get(url)
    # Vérifier si la requête a réussi
    if response.status_code != 200:
        print(f"Erreur : Impossible de récupérer la page (statut {response.status_code})")
        return

    # Parser le contenu HTML
    soup = BeautifulSoup(response.content, 'html.parser')

    # Extraire les données
    citations = []
    for quote in soup.find_all('div', class_='quote'):
        texte = quote.find('span', class_='text').get_text()
        auteur = quote.find('small', class_='author').get_text()
        tags = [tag.get_text() for tag in quote.find_all('a', class_='tag')]
        citations.append({
            'texte': texte,
            'auteur': auteur,
            'tags': tags
        })

    # Afficher les données extraites
    for citation in citations:
        print(f"Citation : {citation['texte']}")
        print(f"Auteur : {citation['auteur']}")
        print(f"Tags : {', '.join(citation['tags'])}")
        print('-' * 40)

# URL de la page à scraper
url = "http://quotes.toscrape.com/"
recuperer_donnees(url)
```
